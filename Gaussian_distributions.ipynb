```python
#Generate a random dataset with normal distribution
import numpy as np
rand_data = np.random.normal(loc=0, scale=1, size=500)
rand_data
rand_data.shape
```

```python
import statistics as st
#Verify the mean and standard deviation of the created random dataset
# mean
st.mean(rand_data)
# standard deviation
st.stdev(rand_data)
```

```python
import statistics as st
import matplotlib.pyplot as plt
from scipy.stats import norm
#h--> return a tuples h[0]-->array or list of arrays ,h[1]-->nbins + 1
#If an integer is given, bins + 1 bin edges are calculated and returned
h = plt.hist(rand_data, bins=30, histtype='bar', density=1, ec='k')
#pdf-->Probability density function.
plt.plot(h[1], norm.pdf(h[1], st.mean(rand_data), st.stdev(rand_data)), color='r', linewidth=2)
plt.title(r'Mean=%.4f, Std Dev=%.4f' %(st.mean(rand_data), st.stdev(rand_data)))
plt.show()
```

```python
#Check the standard normal distribution of the randomly generated data
import statsmodels.api as sm
#Create a QQ plot
sm.qqplot(rand_data, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.show()
# graphic interpretation-->The QQ plot suggests that the generated random data is normally distributed (data plotted on a straight line)
```
